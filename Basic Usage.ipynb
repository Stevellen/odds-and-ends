## Sandpiles

Sandpiles are fun little automata that are capable of displaying some pretty complex interactions and intricate patterns. The Sandpile class is designed to let you explore how these objects change and evolve with different starting sizes and patterns. Creating sandpiles is simple enough, you just need to provide the size and maximum height of a given pile (cell). Note that Sandpiles are represented by 2d numpy arrays with an override of the + operator to manage the toppling behavior. 

You can make some sand like this:


```python
(sand := Sandpile(shape=(11,11), max_height=3))
```

As you can see, spawning a Sandpile without providing a set of points makes a random pile of sand. If you want to look at the sand, you can do so with the show() method

```python
sand.show()
```

Yep, that's sand. What's interesting about these is that you can just pour sand on some sand. When a given cell reaches the maximum height you provided, it will topple. In the current implementation, the only toppling behavior dumps extra sand into the cell's cardinal neighbors (top, right, bottom, left). This toppling has a ripple effect, which is cool because there are times when even a single grain of sand can cause drastic changes in the topography of the sanpile. Adding can be done in a couple of ways. The simplest is using the + operator

```python
(sand+sand).show()
```

See? Such a drastic change from just putting more sand on some sand. We can also animate the sand with the watch_sand method. 

```python
anim = sand.watch_sand(sand.drip(sand.shape[0]))
```
